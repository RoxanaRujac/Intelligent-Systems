### 4. gradient descent

```python
def predict(x, w, b): 
    p = np.dot(x, w) + b     
    return p    
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0.0
    for i in range(m):                                
        f_wb_i = np.dot(x[i], w) + b           
        cost = cost + (f_wb_i - y[i])**2      
    cost = cost / (2 * m)                         
    return cost
```

```python
def compute_gradient(x, y, w, b): 
    m,n = x.shape          
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):                             
        err = (np.dot(x[i], w) + b) - y[i]   
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * x[i, j]    
        dj_db = dj_db + err                        
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m                                
        
    return dj_db, dj_dw
```

```python
import copy

def gradient_descent(x, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    J_history = []
    w = copy.deepcopy(w_in)  #avoid modifying global w within function
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_db,dj_dw = gradient_function(x, y, w, b)

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               
        b = b - alpha * dj_db              
      
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append(cost_function(x, y, w, b))

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i % math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")
        
    return w, b, J_history
```

```python
# initialize parameters
n = len(x_train[0])
initial_w = np.zeros(n)
initial_b = 0.

# some gradient descent settings
iterations = 10
alpha = 7.0e-1

# run gradient descent 
w_final, b_final, J_hist = gradient_descent(x_train, y_train, initial_w, initial_b,
                                                    compute_cost, compute_gradient, 
                                                    alpha, iterations)

print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")

m,_ = x_train.shape
for i in range(m):
    print(f"prediction: {np.dot(x_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")
    
```

```python
# plot cost versus iteration 
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4))
ax1.plot(J_hist)
ax2.plot(100 + np.arange(len(J_hist[100:])), J_hist[100:])
ax1.set_title("Cost vs. iteration");  ax2.set_title("Cost vs. iteration (tail)")
ax1.set_ylabel('Cost')             ;  ax2.set_ylabel('Cost') 
ax1.set_xlabel('iteration step')   ;  ax2.set_xlabel('iteration step') 
plt.show()
```

### 5. r squared score

```python
def r2_score(y, y_):
    num = np.sum((y - y_)**2)
    denom = np.sum((y - y.mean())**2)

    score = 1 - num / denom
    return score
```

```python
y_predicted_train = [np.dot(i, w_final) + b_final for i in x_train]
r2_score(y_train, y_predicted_train).item()
```

```python
y_predicted_val = [np.dot(i, w_final) + b_final for i in x_validation]
r2_score(y_validation, y_predicted_val)
```

### 6. z-score normalization

```python
def z_score_normalize(x):
    mu     = np.mean(x, axis=0)                 # mu will have shape (n,)
    sigma  = np.std(x, axis=0)                  # sigma will have shape (n,)
    x_norm = (x - mu) / sigma      

    return (x_norm, mu, sigma)
```

```python
x_norm, x_mu, x_sigma = z_score_normalize(x_train)
```

```python
w_norm, b_norm, hist = gradient_descent(x_norm, y_train, np.zeros(x_norm.shape[1]), 0., compute_cost, compute_gradient, 1.0e-1, 1000)
```

```python
y_pred_train = [np.dot(x, w_norm) + b_norm for x in x_norm]
r2_score(y_train, y_pred_train).item()
```

```python
x_val_norm = (x_val - x_mu) / x_sigma
y_pred_val = [np.dot(x, w_norm) + b_norm for x in x_val_norm]
r2_score(y_val, y_pred_val)
```
